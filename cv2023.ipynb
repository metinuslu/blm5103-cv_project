```python
import requests
```

```python
df = pd.read_csv("/kaggle/input/movie-genre-from-its-poster/MovieGenre.csv", encoding = "ISO-8859-1")
```

```python
df.columns = [col.lower() for col in df.columns.tolist()]
```

```python
df.head()
```

```python
# df is dataframe with films and poster links # no easy way to match posters and films
for index, row in df.iterrows():
    film = row["title"]
    poster = row["poster"]
    file_name = "/kaggle/working/moviesposters/" + film + ".jpg" # you can change the extension as you like
    try:
        print("Downloading file: %s" % file_name)
        r = requests.get(poster, stream=True)
        with open(file_name, 'wb') as f:
            for chunk in r:
                f.write(chunk)
    except:
        continue
```
